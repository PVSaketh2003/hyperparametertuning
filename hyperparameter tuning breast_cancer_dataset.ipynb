```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

```python
#importing in pandas
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
data_frame
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
#data distribution of target variable
```

```python
data_frame['label'].value_counts()
```

```python
#seperate features and target
```

```python
X=data_frame.drop('label',axis='columns')
y=data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X=np.asarray(X)
y=np.asarray(y)
```

## GRIDSEARCHCV

```python
#step1 - load svc model
```

```python
model=SVC()
```

```python
#step 2 - create dictionary for hyperparameters
```

```python
parameters={'kernel':['linear','poly','rbf','sigmoid'],
            'C':[1,5,10,20]}
```

```python
#step 3 - now do grid searching
```

```python
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#step 4 - fit the data
```

```python
classifier.fit(X,y)
```

```python
# step 5 - check for best possible case
```

```python
classifier.cv_results_
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
#step 6 print for best parameter
```

```python
best_parameters=classifier.best_params_
print(best_parameters)
```

```python
#step 7 - print for high score
```

```python
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results.head()
```

```python
# step 8 - last step
```

```python
grid_search_result=results[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

```python
model=SVC( C=10.0,kernel='linear')
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
#step1 - load svc model
#step 2 - create dictionary for hyperparameters
#step 3 - now do grid searching
#step 4 - fit the data
#step 5 - check for best possible case
#step 6 print for best parameter
#step 7 - print for high score
#step 8 - last step bringing to dataFrame
```

## RandomizedSearchCV

```python
model=SVC()
```

```python
parameters={'kernel':['linear','rbf','poly','sigmoid'],
             'C':[1,5,10,20]}
```

```python
classifier=RandomizedSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_ #results
```

```python
classifier.best_params_ #best parameters
```

```python
classifier.best_score_ #highest Accuracy
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results.head()
```

```python
randomized_search_results=results[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_results #c=10,kernel=linear
```

```python
model=SVC(kernel='linear',C=10.0)
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
